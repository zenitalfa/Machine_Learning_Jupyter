## Özelleştirme

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizisi = np.linspace(0,10,20)
```

```python
numpyDizisi
```

```python
numpyDizisi2 = numpyDizisi**3
```

```python
numpyDizisi2
```

```python
%matplotlib inline
plt.plot(numpyDizisi,numpyDizisi2,"go")
plt.xlabel("Numpy Dizisi 1")
plt.ylabel("Numpy Dizisi 2")
```

## SUBPLOT 

```python
## Subplot

plt.subplot(1,2,1) # 1 sıra ve 2 grafik olacak ve şu an 1. grafik çiziliyor
plt.plot(numpyDizisi,numpyDizisi2,"bo")
plt.subplot(1,2,2) # 2. grafik çiziliyor
plt.plot(np.array(range(20)), numpyDizisi2,"g*")
```

## Figure CLASS'I OLUŞTURMA

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2,0.5,0.9,0.9]) # axisler liste olarak verilmelidir

figureAxes.plot(numpyDizisi, numpyDizisi2)
figureAxes.set_xlabel("Numpy Dizisi 1")
figureAxes.set_ylabel("Numpy Dizisi 2")
figureAxes.set_title ( "Dizilerin grafikleri")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.1,0.4,0.5,0.4])

eksen1.plot(numpyDizisi,numpyDizisi2)
eksen1.set_xlabel("Numpy Dizisi 1")
eksen1.set_ylabel("Numpy Dizisi 2 ")
eksen1.set_title("Ana grafik")

eksen2.plot((np.array(range(20)), numpyDizisi2))
eksen2.set_xlabel("Range 20")
eksen2.set_ylabel("Numpy Dizisi 2")
eksen2.set_title("Küçük grafik başlık")
```
