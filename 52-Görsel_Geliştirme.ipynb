## Görsel Geliştirme

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## LEGEND kullanımı 

```python
yeniFigure = plt.figure()

## yeniFigure = plt.figure(figsize= (10,9)) şeklinde grafiğin büyüklüğü verilebilir, eksen kullanmadan ve içerisi tuple

eksen1 = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
#eksen2 = yeniFigure.add_axes([0.3,0.2,0.9,0.9])

numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi2 = numpyDizisi1**2

eksen1.plot(numpyDizisi1,numpyDizisi2, label = "Numpy Dizisi1**2")
eksen1.plot(numpyDizisi1, label= "Numpy Dizisi1")
eksen1.legend()

## eğer legend ibaresinin lokasyonunu belirlemek istersek
## eksen1.legend(loc=1 ya da loc=2 ya da loc= 3 gibi bazı sayılar verilebilir)
```

## DPİ (Dots per inches) inch başına pixel yani çözünürlük girişi yapabiliriz

```python
yeniFigure = plt.figure(dpi = 200)

## yeniFigure = plt.figure(figsize= (10,9)) şeklinde grafiğin büyüklüğü verilebilir, eksen kullanmadan ve içerisi tuple
## dpi değerleriyle oynayarak da çözünürlük değiştirilebilir
eksen1 = yeniFigure.add_axes([0.1,0.1,0.9,0.9])
#eksen2 = yeniFigure.add_axes([0.3,0.2,0.9,0.9])

numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi2 = numpyDizisi1**2

eksen1.plot(numpyDizisi1,numpyDizisi2, label = "Numpy Dizisi1**2")
eksen1.plot(numpyDizisi1, label= "Numpy Dizisi1")
eksen1.legend()

## eğer legend ibaresinin lokasyonunu belirlemek istersek
## eksen1.legend(loc=1 ya da loc=2 ya da loc= 3 gibi bazı sayılar verilebilir)
```

## Figure dosyasını kaydetmek için yeniFigure.savefig komutu

```python
yeniFigure.savefig("figure dosyası.jpeg", dpi = 200)

# dpi burada da kullanılabilir
```
