## Tensorflow kütüphaneleri ve model yüklenmesi

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout

## Regresyondan farklı olarak burada Activation ve Dropout gibi

from tensorflow.keras.callbacks import EarlyStopping
```

### EarlyStopping ve Dropout, overfittingi çözmek için kullanacağız

```python
x_train.shape
```

```python
x_test.shape
```

## Model Oluşturma

```python
model= Sequential()

model.add(Dense(units= 30,activation="relu"))
model.add(Dense(units= 15,activation="relu"))
model.add(Dense(units= 15,activation="relu"))
#model.add(Dense(units= 30,activation="relu"))

## genel kurala göre kaç tane feature varsa o kadar giriş nöronu olması önerilir
model.add(Dense(units=1, activation="sigmoid"))

## Sigmoid, çıkışında 0 veya 1 verir
## Classification problemlerinde kullanılabilir

model.compile(optimizer= "adam", loss="binary_crossentropy")
```

## Train

```python
model.fit(x=x_train, y= y_train, epochs=700 , validation_data= (x_test,y_test),verbose=1)

## Burada bilerek overfitting yaptırıyoruz
```

```python
model.history.history
```

## Overfitting'i engellemek için Earlystopping kullanılır

```python
model= Sequential()

model.add(Dense(units= 30,activation="relu"))
model.add(Dense(units= 15,activation="relu"))
model.add(Dense(units= 15,activation="relu"))
#model.add(Dense(units= 30,activation="relu"))

## genel kurala göre kaç tane feature varsa o kadar giriş nöronu olması önerilir
model.add(Dense(units=1, activation="sigmoid"))

## Sigmoid, çıkışında 0 veya 1 verir
## Classification problemlerinde kullanılabilir

model.compile(optimizer= "adam", loss="binary_crossentropy")
```

```python
earlyStopping= EarlyStopping(monitor="val_loss", mode= "min",verbose=1,patience=25)

## SHIFT+TAB ile dokümantasyonuna bak
```

```python
model.fit(x= x_train, y= y_train, epochs= 700, validation_data=(x_test,y_test), 
          verbose=1, callbacks=[earlyStopping])
```

```python
modelKaybi= pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
## Gerçekten de modelde bir iyilşme söz konusudur
```

## Dropout

```python
## Dropout, training sırasında bazı layerları random olarak atarak overfittingi engeller ve
## daha iyi bir model sunmaya çalışıt
```

```python
## Model yeniden alınır
```

```python
model= Sequential()


## Dropout her katmandan sonra eklenir, son katman hariç
## 0 ve 1 arasında bir değer alır ancak 0 etkisi yok ve 1 de etkisi maksimum olur
## 0.5 değerinden sonra da uyarı verir

model.add(Dense(units= 30,activation="relu"))
model.add(Dropout(0.6))


model.add(Dense(units= 15,activation="relu"))
model.add(Dropout(0.6))


model.add(Dense(units= 15,activation="relu"))
model.add(Dropout(0.6))
#model.add(Dense(units= 30,activation="relu"))

## genel kurala göre kaç tane feature varsa o kadar giriş nöronu olması önerilir
model.add(Dense(units=1, activation="sigmoid"))

## Sigmoid, çıkışında 0 veya 1 verir
## Classification problemlerinde kullanılabilir

model.compile(optimizer= "adam", loss="binary_crossentropy")
```

```python
model.fit(x = x_train, y= y_train, epochs= 700, validation_data= (x_test, y_test), 
          verbose=1, callbacks=[earlyStopping])

```

```python
modelLoss= pd.DataFrame(model.history.history)
```

```python
modelLoss.plot()
```

```python
tahminlerimiz= model.predict_classes(x_test)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix

## Confusion matrix, classificaition accuracy hakkında bilgi verir
```

```python
## y_test ve tahminlerimiz'i kıyaslamak için classification report çağırılır
print(classification_report(y_test, tahminlerimiz))
```

```python
print(confusion_matrix(y_test,tahminlerimiz))

## Burada 7 değeri çok önemli
## 7 değer yanlış bulmuş, geri kalanları doğru bulmuştur
```
