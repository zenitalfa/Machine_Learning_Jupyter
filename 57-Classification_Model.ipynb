```python
model= Sequential()

model.add(Dense(units= 30,activation="relu"))
model.add(Dense(units= 15,activation="relu"))
model.add(Dense(units= 15,activation="relu"))
#model.add(Dense(units= 30,activation="relu"))

## genel kurala gÃ¶re kaÃ§ tane feature varsa o kadar giriÅŸ nÃ¶ronu olmasÄ± Ã¶nerilir
model.add(Dense(units=1, activation="sigmoid"))

## Sigmoid, Ã§Ä±kÄ±ÅŸÄ±nda 0 veya 1 verir
## Classification problemlerinde kullanÄ±labilir

model.compile(optimizer= "adam", loss="binary_crossentropy")
```

## Dropout

```python
## Dropout, training sÄ±rasÄ±nda bazÄ± layerlarÄ± random olarak atarak overfittingi engeller ve
## daha iyi bir model sunmaya Ã§alÄ±ÅŸÄ±t
```

```python
## Model yeniden alÄ±nÄ±r
```

```python
model= Sequential()


## Dropout her katmandan sonra eklenir, son katman hariÃ§
## 0 ve 1 arasÄ±nda bir deÄŸer alÄ±r ancak 0 etkisi yok ve 1 de etkisi maksimum olur
## 0.5 deÄŸerinden sonra da uyarÄ± verir

model.add(Dense(units= 30,activation="relu"))
model.add(Dropout(0.6))


model.add(Dense(units= 15,activation="relu"))
model.add(Dropout(0.6))


model.add(Dense(units= 15,activation="relu"))
model.add(Dropout(0.6))
#model.add(Dense(units= 30,activation="relu"))

## genel kurala gÃ¶re kaÃ§ tane feature varsa o kadar giriÅŸ nÃ¶ronu olmasÄ± Ã¶nerilir
model.add(Dense(units=1, activation="sigmoid"))

## Sigmoid, Ã§Ä±kÄ±ÅŸÄ±nda 0 veya 1 verir
## Classification problemlerinde kullanÄ±labilir

model.compile(optimizer= "adam", loss="binary_crossentropy")
```

```python
model.fit(x = x_train, y= y_train, epochs= 700, validation_data= (x_test, y_test), 
          verbose=1, callbacks=[earlyStopping])

```

```python
modelLoss= pd.DataFrame(model.history.history)
```

```python
modelLoss.plot()
```

```python
tahminlerimiz= model.predict_classes(x_test)
```

```python
tahminlerimiz
```

```python
from sklearn.metrics import classification_report, confusion_matrix

## Confusion matrix, classificaition accuracy hakkÄ±nda bilgi verir
```

```python
## y_test ve tahminlerimiz'i kÄ±yaslamak iÃ§in classification report Ã§aÄŸÄ±rÄ±lÄ±r
print(classification_report(y_test, tahminlerimiz))
```

```python
print(confusion_matrix(y_test,tahminlerimiz))

## Burada 7 deÄŸeri Ã§ok Ã¶nemli
## 7 deÄŸer yanlÄ±ÅŸ bulmuÅŸ, geri kalanlarÄ± doÄŸru bulmuÅŸtur
```
