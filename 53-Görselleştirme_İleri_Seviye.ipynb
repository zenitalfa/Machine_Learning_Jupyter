## Görselleştirme İleri Seviye

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
numpyDizisi2 = numpyDizisi1**2
```

```python
(Figure, eksen) = plt.subplots()
eksen.plot(numpyDizisi1,numpyDizisi2, "g")
eksen.plot(numpyDizisi1, "m", label = "Numpy dizisi")
plt.legend()
```

## RGB kullanılabilir, rgb hex code bulunur color = "#188DDB"

```python
(Figure, eksen) = plt.subplots()
eksen.plot(numpyDizisi1,numpyDizisi2, color= "#188DDB")
eksen.plot(numpyDizisi1, "m", label = "Numpy dizisi")
plt.legend()
```

## Eksen çizigilerini değiştirme

```python
(yeniFigure,yeniEksen) = plt.subplots()

yeniEksen.plot(numpyDizisi1,numpyDizisi2, color = "#030C12", alpha = 1) 
```

## ÇİZGİ KALINLIĞI

```python
(yeniFigure, eksen) = plt.subplots()
eksen.plot(numpyDizisi1, numpyDizisi1+2, color = "blue", linewidth = 10)
eksen.plot(numpyDizisi1, numpyDizisi1+ 8, color = "red", linewidth= 4)
```

## ÇİZGİ STİLİ ve marker

```python
## linestyle = "--" gibi bir metod ile bu çizgi stili değiştirilebilir
(yeniFigure, eksen) = plt.subplots(dpi = 100)
eksen.plot(numpyDizisi1, numpyDizisi1+2, color = "blue", linewidth = 10, linestyle = "-.")
eksen.plot(numpyDizisi1, numpyDizisi1+12, color = "red", linewidth= 4, linestyle =":" )

## Marker
eksen.plot(numpyDizisi1, numpyDizisi1+5, color= "#207769" , linewidth = 7, linestyle = "--", 
           markersize= 10, markerfacecolor= "red",marker = "*")
```
