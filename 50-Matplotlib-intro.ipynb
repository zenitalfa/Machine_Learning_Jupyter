## Matplotlib Giriş

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# Jupiter notebook için satır içerisinde grafik gösterilmesini sağlar
```

```python
yasListesi = [10,32,13,45,67,9,18,23,51,83]
kiloListesi = [42,84,50,78,90,23,64,77,49,102]
```

```python
len(kiloListesi)
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

## GRAFİK

```python
plt.plot(numpyYasListesi, numpyKiloListesi, "m")
plt.xlabel("Yaş")
plt.ylabel("Kilo")
plt.title("Yaş-Kilo Grafiği")
## Başka editörler için plt.show() yazmamız gerekir.
## Yukarıda matplotlib inline yazdığımız için buna gerek yok şu anlık
```
